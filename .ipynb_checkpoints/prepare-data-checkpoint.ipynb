```python
import pandas as pd
df=pd.read_csv('data/train.csv',sep="\t")
df.head()
```

```python
df.Sentiment.describe()
```

```python
df.dtypes
```

```python
df.Phrase=df.Phrase.str.lower()
df.head()
```

```python
df.Phrase = df.Phrase.str.replace('[^\w\s]','')
df.head()
```

```python
singleMask = df.Phrase.str.match(r'\A[\w-]+\Z')
singleMask[0:5]
```

```python
for x in range(0,singleMask.size):
    if singleMask[x]==True:
        singleMask[x]=False
    else:
        singleMask[x]=True
singleMask[0:5]
```

```python
df=df[singleMask]
df.head(n=40)
```

```python
import numpy as np
def change_nan(x):
    if pd.isnull(x):
        return np.nan
    
    if len(x) == 0:
        return np.nan
    return x
```

```python
df['Phrase'] = df.Phrase.map(change_nan)
df=df.dropna()
df.head(n=40)
```

```python
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
```

```python
def stop_erase(x):
    proper_words = []
    words = x.split()
    for word in words:
        if not word in stopwords_set:
           proper_words.append(word)
        
    return " ".join(proper_words)
```

```python
df.Phrase=df.Phrase.map(stop_erase)
```

```python
df.head(n=30)
```

```python
df['Phrase'] = df.Phrase.map(change_nan)
df=df.dropna()
df.head(n=40)
```

```python
df.dtypes
```

```python
df = df.reset_index(drop=True)
df.head(n=30)
```

```python
df.to_csv("data/cleaneddf.csv", sep='\t')
```
