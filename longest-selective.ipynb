```python
import pandas as pd
df=pd.read_csv('data/train.csv',sep="\t").copy(deep=True)
path='C:\\Users\\efear\\Documents\\GitHub\\Sentiment-Analysis-on-Movie-Reviews\\data\\longest-selective\\'
df.head()
```

```python
msk=[]
tmp=0

for x in df.SentenceId:
    if x !=tmp:
        msk.append(True)
        tmp=x
    else:
        msk.append(False)
        
msk=pd.Series(msk)
df=df[msk]
df.head()
```
